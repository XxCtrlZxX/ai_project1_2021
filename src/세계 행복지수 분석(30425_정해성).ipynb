```python
# 데이터 가져오기
import pandas as pd

hp_df_2021 = pd.read_csv('world-happiness-report-2021.csv')
hp_df = pd.read_csv('world-happiness-report.csv')
```

## 3. 탐색적 데이터 분석 과정

#### 표를 활용하여 분석하기

---
### 속성 설명
1. Country name : 국가 이름
2. Regional indicator: 해당 국가가 속해있는 대륙
3. Ladder score : 행복 지수
4. Logged GDP per capita : 1인당 소득
5. Social support : 사회적 지지
6. Healthy life expectancy : 기대수명
7. Freedom to make life choices : 자유도
8. Generosity : 관용 의식
9. Perceptions of corruption : 부패 인식

```python
# pandas 활용
print('Dataframe 크기 :', hp_df_2021.shape)
hp_df_2021.head()
```

#### 필요한 칼럼만 추출하여 저장 

```python
attr = ['Country name', 'Regional indicator', 'Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity']
happiness_df_2021 = hp_df_2021[attr]
happiness_df_2021.head()
```

#### 그래프를 활용하여 분석하기

```python
# matplotlib, seaborn 등 활용
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic') # 한글 폰트 설정
plt.rc('axes', unicode_minus=False)   # 유니코드에서  음수 부호설정

fig, ax = plt.subplots(3, 2, figsize=(10, 10))

ax[0, 0].plot(happiness_df_2021['Ladder score'][:50], color='r')
ax[0, 1].plot(happiness_df_2021['Logged GDP per capita'][:50])
ax[1, 0].plot(happiness_df_2021['Social support'][:50])
ax[1, 1].plot(happiness_df_2021['Healthy life expectancy'][:50])
ax[2, 0].plot(happiness_df_2021['Freedom to make life choices'][:50])
ax[2, 1].plot(happiness_df_2021['Generosity'][:50])

ax[0, 0].set_title('행복지수')
ax[0, 1].set_title('GDP')
ax[1, 0].set_title('사회적 지지')
ax[1, 1].set_title('기대수명')
ax[2, 0].set_title('자유도')
ax[2, 1].set_title('관대함')

fig.tight_layout()
plt.show()
```

#### 대륙 별 평균 행복지수

```python
import numpy as np

continent_mean = happiness_df_2021.groupby('Regional indicator').mean()['Ladder score']

continent_names = ['중,동부 유럽', '독립 국가 연합', '동아시아', '라틴아메리카', '중,북부 아프리카', '북아메리카', '남아시아', '동남아시아', '아프리카', '서유럽']

continent_mean.plot.barh(color='g')
plt.yticks(range(10), continent_names)
plt.ylabel('대륙')
plt.xlabel('평균 행복지수')
plt.show()
```

#### 우리나라 연도별 행복지수 외 5가지 요소의 수치 변화 분석

```python
# 데이터 프레임 칼럼 이름 충돌 방지 위함
# {'바꿀 이름': '바뀔 이름'}

renameList = {
    'Life Ladder': 'Ladder score',
    'Log GDP per capita': 'Logged GDP per capita',
    'Healthy life expectancy at birth': 'Healthy life expectancy'
}
```

```python
kor = hp_df[hp_df['Country name'] == 'South Korea'] # 우리나라만 추출

kor_df = [
    kor[kor['year'] == 2010 + i].rename(columns=renameList)
    for i in range(11)
] # 연도별 추출(2010 ~ 2020)과 동시에 칼럼 이름 변경

kor_df.append(hp_df_2021[hp_df_2021['Country name'] == 'South Korea']) # 2021년 따로 추가
```

```python
factor = attr[2:]
years = [2010 + i for i in range(12)]

titles = ['연도별 행복지수', '연도별 GDP', '연도별 사회적 지지']

# for f in factor:
#     for k in kor_df:
#         print(k[f].values[0])

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for n, ax in enumerate(axs):
    v = list()
    for k in kor_df:
        v.append(k[factor[n]].values[0]) # 값을 하나하나 리스트에 대입함
        
    ax.plot(v)
    ax.set_title(titles[n])
    ax.set_xticks(range(12))
    ax.set_xticklabels(years, rotation=40)

fig.tight_layout()
plt.show()
```
