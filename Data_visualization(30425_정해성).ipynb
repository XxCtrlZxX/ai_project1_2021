### 1) 그림의 구조
- matplotlib가 그리는 그림은 Figure 객체, Axes 객체, Axis 객체 등으로 구성된다.
- Figure 객체는 한 개 이상의 Axes 객체를 포함하고 Axes 객체는 다시 두 개 이상의 Axis 객체를 포함한다.
- Figure는 그림이 그려지는 캔버스나 종이를 뜻하고 Axes는 하나의 플롯, 그리고 Axis는 가로축이나 세로축 등의 축을 뜻한다.

### 2) 선 그래프(line plot)

```python
import matplotlib.pyplot as plt # 서브 라이브러리
```

```python
# 한글 폰트 지정하기()
plt.rc('font', family='Malgun Gothic')
```

- y축에 해당하는 데이터는 반드시 입력해야 한다.
- x축에 해당하는 데이터를 입력하지 않으면 0부터 시작되는 정수로 y축 데이터의 개수에 맞춰진다.
- 그래프를 보여주려면 show() 함수를 실행한다.

```python
plt.plot([1, 3, 6, 9, 13])
plt.show()
```

```python
plt.plot([3, 4, 5, 6, 7, 8], [13, 16, 21, 25, 29, 31])
plt.show()
```

```python
plt.plot([3, 4, 5, 6, 7, 8], [13, 16, 21, 25, 29, 31], 'ro-')
plt.show()
```

```python
plt.plot([3, 4, 5, 6, 7, 8], [13, 16, 21, 25, 29, 31], 'ro-')
plt.axis([2.5, 8.5, 10, 40]) # [xmin, xmax, ymin, ymax] x, y축의 범위 설정
plt.show()
```

```python
plt.plot([3, 4, 5, 6, 7, 8], [13, 16, 21, 25, 29, 31], 'ro-')
plt.axis([2.5, 8.5, 10, 40]) # [xmin, xmax, ymin, ymax] x, y축의 범위 설정
plt.xlabel('월')
plt.ylabel('온도')
plt.show()
```

#### 수식 활용

```python
x = range(0, 100)
y = [v*v for v in x]
plt.plot(x, y)
plt.show()
```

#### numpy 라이브러리 활용

```python
# y = x, y = x**2, y = x**3
# 색 : r g b k y
# 마커(데이터의 위치를 나타내는 기호) . , ^ s o + *
# 선 - -- -.
import numpy as np

x_range = np.arange(0, 5, 0.2)

plt.plot(x_range, x_range, 'r^-')
plt.plot(x_range, x_range**2, 'gs-')
plt.plot(x_range, x_range**3, 'b--')
plt.axis([-0.5, 5.5, -0.5, 5.5])
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('여러 종류 그래프')
plt.show()
```

#### pandas 라이브러리 활용

```python
from pandas import Series, DataFrame

s1 = Series([3, 10, 34, 6, 15])
s1.plot()
plt.show()
```
